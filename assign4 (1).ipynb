```python
#!pip install nltk gensim scipy

# Import required libraries
import nltk
import string
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Download tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Input Text Corpus
# -----------------------------
text = """
I love machine learning.
Machine learning is very useful.
I love coding in python.
Python is great for machine learning.
"""
```

```python
 #Preprocessing
# -----------------------------
text = text.lower()

text = text.translate(str.maketrans('', '', string.punctuation))

sentences = text.split('\n')

tokenized_data = []
for sentence in sentences:
    words = word_tokenize(sentence)
    if len(words) > 0:
        tokenized_data.append(words)

print("Tokenized Data:")
print(tokenized_data)
```

```python
# Train Word2Vec Model
# -----------------------------
model = Word2Vec(
    sentences=tokenized_data,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4
)
```

```python
# Word Vectors
# -----------------------------
print("\nWord Vector for 'machine':")
print(model.wv['machine'])

# Similarity Check
# -----------------------------
similarity = model.wv.similarity('machine', 'learning')
print("\nSimilarity between 'machine' and 'learning':")
print(similarity)
```

```python
print("\nWords similar to 'machine':")
print(model.wv.most_similar('machine'))
```
