```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "best laptop for programming",
    "top smartphones under budget",
    "how to learn machine learning",
    "best books for python coding",
    "latest technology news",
    "how to fix login issue",
    "payment failed but money deducted"
]
```

```python
import re

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    words = text.split()
    return words
```

```python
# Create vocabulary
vocab = set()
for doc in documents:
    vocab.update(preprocess(doc))

vocab = list(vocab)

# Assign random vectors (simulate embeddings)
embedding_dim = 50
word_vectors = {word: np.random.rand(embedding_dim) for word in vocab}
```

```python
def sentence_vector(sentence):
    words = preprocess(sentence)
    vectors = []
    
    for word in words:
        if word in word_vectors:
            vectors.append(word_vectors[word])
    
    if len(vectors) == 0:
        return np.zeros(embedding_dim)
    
    return np.mean(vectors, axis=0)
```

```python
doc_vectors = [sentence_vector(doc) for doc in documents]
```

```python
def semantic_search(query):
    query_vec = sentence_vector(query)
    
    similarities = []
    
    for vec in doc_vectors:
        sim = cosine_similarity([query_vec], [vec])[0][0]
        similarities.append(sim)
    
    # Get best match
    best_index = np.argmax(similarities)
    
    return documents[best_index], similarities[best_index]
```

```python
query = "best laptop for coding"

result, score = semantic_search(query)

print("Query:", query)
print("Most Relevant Result:", result)
print("Similarity Score:", score)
```
