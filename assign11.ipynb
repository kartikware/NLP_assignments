```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
from nltk.chunk import RegexpParser

# Download required resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
paragraph = """Natural language processing techniques are widely used in artificial intelligence applications. 
The intelligent system extracts meaningful information from large datasets."""
```

```python
sentences = sent_tokenize(paragraph)
```

```python
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
chunk_parser = RegexpParser(grammar)
```

```python
noun_phrases = []

for sentence in sentences:
    # Word Tokenization
    words = word_tokenize(sentence)
    
    # POS Tagging
    tagged_words = pos_tag(words)
    
    # Chunk Parsing
    tree = chunk_parser.parse(tagged_words)
    
    # Extract NP
    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            np = " ".join(word for word, tag in subtree)
            noun_phrases.append(np)

noun_phrases
```
