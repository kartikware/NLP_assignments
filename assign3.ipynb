```python
import nltk
nltk.download('punkt_tab')
```

```python
#corpus
text = "I love machine learning. I love natural language processing. Machine learning is fun. I am just kidding LOL."
print("Original Text:\n", text)
```

```python
#preprocessing
from nltk.tokenize import word_tokenize
import string

text = text.lower()

text = text.translate(str.maketrans('', '', string.punctuation))

tokens = word_tokenize(text)

print("Tokens:\n", tokens)
```

```python
#generate n gram
from nltk.util import ngrams

# Unigrams
unigrams = list(ngrams(tokens, 1))

# Bigrams
bigrams = list(ngrams(tokens, 2))

# Trigrams
trigrams = list(ngrams(tokens, 3))

print("Unigrams:\n", unigrams[:18])
print("\nBigrams:\n", bigrams[:18])
print("\nTrigrams:\n", trigrams[:18])
```

```python
#freq count
from collections import Counter

unigram_freq = Counter(unigrams)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

print("Top Unigrams:\n", unigram_freq.most_common(5))
print("\nTop Bigrams:\n", bigram_freq.most_common(5))
print("\nTop Trigrams:\n", trigram_freq.most_common(5))
```

```python
#probability of bigram
bigram_prob = {}

for (w1, w2), count in bigram_freq.items():
    bigram_prob[(w1, w2)] = count / unigram_freq[(w1,)]

for k, v in list(bigram_prob.items())[:5]:
    print(f"{k} -> {v}")
```

```python
#predict next word
def predict_next_word(word):
    candidates = {}

    for (w1, w2), prob in bigram_prob.items():
        if w1 == word:
            candidates[w2] = prob

    if candidates:
        return max(candidates, key=candidates.get)
    else:
        return "No prediction available"

# Test
print("Next word after 'machine':", predict_next_word("machine"))
print("Next word after 'love':", predict_next_word("love"))
```

```python
#trigram prediction
trigram_prob = {}

for (w1, w2, w3), count in trigram_freq.items():
    trigram_prob[(w1, w2, w3)] = count / bigram_freq[(w1, w2)]

def predict_next_word_trigram(w1, w2):
    candidates = {}

    for (a, b, c), prob in trigram_prob.items():
        if a == w1 and b == w2:
            candidates[c] = prob

    if candidates:
        return max(candidates, key=candidates.get)
    else:
        return "No prediction available"

# Test
print("Next word after 'i love':", predict_next_word_trigram("i", "love"))
```
