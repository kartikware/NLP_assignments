```python
# Text Preprocessing
import nltk
import re
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text = """Natural language processing is a field of artificial intelligence.
Machine learning and deep learning are subsets of AI.
Word embeddings capture semantic meaning of words."""
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    sentences = text.split("\n")
    words = [word_tokenize(sent) for sent in sentences]
    return words

processed_data = preprocess(text)
processed_data
```

```python
# Train Word2Vec Model
from gensim.models import Word2Vec

w2v_model = Word2Vec(sentences=processed_data, vector_size=50, window=3, min_count=1)
```

```python
# Test Word2Vec Model
print("Vector for 'learning':\n", w2v_model.wv['learning'])
print("\nSimilar words:\n", w2v_model.wv.most_similar('learning'))
```

```python
# Load GloVe Model
def load_glove(file_path):
    glove_model = {}
    with open(file_path, encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = list(map(float, values[1:]))
            glove_model[word] = vector
    return glove_model

# Make sure glove.6B.50d.txt is in your folder
glove = load_glove("glove.6B.50d.txt")
```

```python
# Test GloVe Model
print("Vector for 'learning':\n", glove['learning'][:10])
```

```python
# Train FastText Model
from gensim.models import FastText

fast_model = FastText(sentences=processed_data, vector_size=50, window=3, min_count=1)
```

```python
# Test FastText Model
print("Vector for 'learning':\n", fast_model.wv['learning'])
print("\nSimilar words:\n", fast_model.wv.most_similar('learning'))
```

```python
# Compare Embedding Models
print("Word2Vec Similar:", w2v_model.wv.most_similar('learning'))
print("FastText Similar:", fast_model.wv.most_similar('learning'))
print("GloVe Vector Sample:", glove['learning'][:5])
```

```python
# Display Results
print("Processed Data:", processed_data)
print("\nWord2Vec Vocabulary:", list(w2v_model.wv.index_to_key))
print("\nFastText Vocabulary:", list(fast_model.wv.index_to_key))
```
