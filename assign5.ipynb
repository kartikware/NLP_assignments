```python
import nltk
nltk.download('punkt_tab')
```

```python
#Create Dataset
# Sample dataset
data = [
    ("book a flight", "booking"),
    ("i want to reserve a ticket", "booking"),
    ("cancel my order", "cancellation"),
    ("i want to cancel my ticket", "cancellation"),
    ("what is the weather today", "weather"),
    ("tell me the weather forecast", "weather"),
    ("hello", "greeting"),
    ("hi there", "greeting")
]

texts = [x[0] for x in data]
labels = [x[1] for x in data]
```

```python
#Text Preprocessing
import string
from nltk.tokenize import word_tokenize

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    return " ".join(tokens)

processed_texts = [preprocess(t) for t in texts]

print(processed_texts)
```

```python
#feature extraction
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_texts)

print("Feature Shape:", X.shape)
```

```python
#train model
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X, labels)
```

```python
#intent prrdictiom
def predict_intent(user_input):
    processed = preprocess(user_input)
    vector = vectorizer.transform([processed])
    prediction = model.predict(vector)
    return prediction[0]

# Test examples
print(predict_intent("book my ticket"))
print(predict_intent("cancel the order"))
print(predict_intent("what is weather tomorrow"))
print(predict_intent("hello bot"))
```

```python
while True:
    user_input = input("Enter your query (type 'exit' to stop): ")
    if user_input.lower() == 'exit':
        break
    print("Predicted Intent:", predict_intent(user_input))
```
