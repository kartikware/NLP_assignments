```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
```

```python
text = "The quick brown fox jumps over the lazy dog. It is a sunny day, so let's buy ice cream "
print("Original Text:\n", text)
```

```python
#lowercase + tokenizaiton
from nltk.tokenize import word_tokenize

# Convert to lowercase
text_lower = text.lower()

# Tokenize
tokens = word_tokenize(text_lower)

print("Tokens:\n", tokens)
```

```python
#pos tagging using NLTK
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger_eng')

pos_tags_nltk = pos_tag(tokens)

print("NLTK POS Tags:\n")
for word, tag in pos_tags_nltk:
    print(f"{word} → {tag}")
```

```python
#pos Tagging using spacy
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

print("SpaCy POS Tags:\n")
for token in doc:
    print(f"{token.text} → {token.pos_}")
```

```python
#comparision
print("Word\t\tNLTK Tag\tSpaCy Tag")
print("--------------------------------------")

for i in range(len(doc)):
    word = doc[i].text
    spacy_tag = doc[i].pos_

    if i < len(pos_tags_nltk):
        nltk_tag = pos_tags_nltk[i][1]
    else:
        nltk_tag = "-"

    print(f"{word}\t\t{nltk_tag}\t\t{spacy_tag}")
```
